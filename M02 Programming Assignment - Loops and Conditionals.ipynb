```python
secret = 6
# the secret number
guess = 6
# the guess being compared to the secret number
if guess < secret:
    print("too low")
elif guess > secret:
    print("too high")
else:
    print("just right")
# the results in whether the guess was higher, lower or equal to the secret number
```

SECTION 4.2

```python
small = False
green = True
# Change "True" and "False" for different outcome of testing

if small and green:
    print("pea")
elif small and not green:
    print("cherry")
elif green and not small:
    print("watermelon")
else:
    print("pumpkin")
```

SECTION 6.2

```python
guess_me = 7
number = 1
# Change the values for different testing outcomes

while not (number > guess_me):
    if number < guess_me:
        print ("too low")
        number += 1
    elif number == guess_me:
        print ("found it!")
        break
    else:
        print("oops")
        break
```
