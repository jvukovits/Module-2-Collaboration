```python
# Jeff Vukovits - Deans List & Honor Checker
# This program will ask for the student's first and last name and their GPA.
# The output will result in whether or not they made the Dean's List or Honor Roll.

last_name = input("What is your last name? (Enter ZZZ to quit)")
while last_name != "ZZZ":
    first_name = input("What is your first name?")
    gpa_score = float(input("What is your GPA?"))
    if gpa_score >= 3.50:
        print ("Congratulations", first_name, last_name, "you have made the Dean's List!")
    else:
        print ("I am sorry", first_name, last_name, "you did not make the Dean's List")
    if gpa_score >= 3.25:
        print ("Congratulations", first_name, last_name, "you have made the Honor Roll!")
    else:
        print ("I am sorry", first_name, last_name, "you did not make the Honor Roll")
    last_name = input("What is your last name? (Enter ZZZ to quit)")
else:
    quit
```
